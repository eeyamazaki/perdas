```python
import os
import glob
import pandas as pd
from google.colab import files
```

```python
os.chdir(r'/content/drive/MyDrive/Projetos/perdas/VD/')
files = glob.glob('*.xlsx')

df = pd.concat([pd.read_excel(file) for file in files]).reset_index()
```

```python
df.head()
```

```python
# Colocando a segunda linha como cabecalho:

df = df.rename(columns=df.iloc[1])
```

```python
# Removendo demais valores de cabecalhos do df:

df = df[df['Descrição'].str.contains('Descrição')==False]
```

```python
# Removendo valores nulos:

df = df.dropna(subset=['Descrição'])
```

```python
df.isnull().sum()
```

```python
dfVRP = df.iloc[:, 1:3]
```

```python
dfVRP = dfVRP.drop_duplicates(subset=['Nº Série'])
```

```python
df.head()
```

```python
# Dropando a ultima coluna:

df = df.drop(['Descrição', 'Volume Total'], axis=1)
```

```python
cols = df.columns[1:]
```

```python
df[cols].head()
```

```python
df = df[cols].set_index('Nº Série').stack().reset_index()
```

```python
df.columns.values[0:] = ['NSerie', 'DataRef', 'VD']
```

```python
df.head()
```

```python
len(df)
```

```python
# Download das tabelas:
from google.colab import files

df.to_csv('VD.csv')
files.download('VD.csv')

dfVRP.to_csv('VRP.csv')
files.download('VRP.csv')
```
